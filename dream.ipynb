```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model_name="gpt-3.5-turbo")
```

```python
urls = [
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-aglamak-ne-anlama-gelir-25196',
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-ask-itirafi-almak-ne-anlama-gelir-6990877',
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-askerlerden-kacmak-ne-anlama-gelir-6990875/',
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-agzindan-kan-geldigini-gormek-ne-anlama-gelir-6990869',
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-agizdan-kan-tukurmek-ne-anlama-gelir-6990866',
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-agactan-incir-toplamak-ne-anlama-gelir-6990864',
    'https://www.milliyet.com.tr/pembenar/ruya-tabirleri/ruyada-aya-gitmek-ne-anlama-gelir-6978525'
]
```

```python
from langchain.document_loaders import UnstructuredURLLoader
loaders = UnstructuredURLLoader(urls=urls)
data = loaders.load()
```

```python
data
```

```python
# Text Splitter
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
chunk_size=1000, chunk_overlap=0, separators=[" ", ",", "\n"]
)


docs = text_splitter.split_documents(data)
```

```python
docs
```

```python
len(docs)
```

```python
import pickle
import faiss
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()
```

```python
embeddings
```

```python
vectorStore_openAI = FAISS.from_documents(docs, embeddings)
```

```python
vectorStore_openAI
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.question_answering import load_qa_chain
from langchain import OpenAI
```

```python
llm=ChatOpenAI(temperature=0, model_name='gpt-3.5-turbo')
```

```python
llm
```

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorStore_openAI.as_retriever())
```

```python
chain({"question": "Rüyada ağlamak ne anlama gelir?"}, return_only_outputs=True)
```
